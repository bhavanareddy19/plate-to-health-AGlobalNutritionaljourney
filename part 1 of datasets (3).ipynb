**considering only necessary data**

```python

datasets = [v01_cnty, v02_cnty, v08_cnty, v09_cnty, v10_cnty, v11_cnty,
            v15_cnty, v16_cnty, v17_cnty, v18_cnty, v22_cnty, v23_cnty, v27_cnty]

filtered_data = []
for df in datasets:
    df = df[df['year'] >= 2010]
    df = df[(df['age'] >= 12.5) & (df['age'] <= 42.5)]
    df = df[(df['gender'] == 0) | (df['gender'] == 1)]
    df = df[(df['region'] == 0) | (df['region'] == 1)]
    filtered_data.append(df)
    print(df.shape)
```

**DROP ALL NULL VALUES**

```python
cleaned_data = []
for df in filtered_data:
    cleaned_df = df.dropna()
    cleaned_data.append(cleaned_df)
    print(cleaned_df.isnull().sum())
```

**Check for and remove duplicate values**

```python
for df in cleaned_data:
    initial_shape = df.shape
    df.drop_duplicates(inplace=True)
    final_shape = df.shape
    print(f"Initial shape: {initial_shape}, Final shape after removing duplicates: {final_shape}")
```

***merging all datasets into one dataset***

```python
import pandas as pd

diet_factors = [
    "Fruits", "Non-starchy vegetables", "Potatoes", "Other starchy vegetables",
    "Beans and legumes", "Nuts and seeds", "Refined grains", "Whole grains",
    "Total processed meats", "Unprocessed red meats", "Total seafoods", "Eggs",
    "Cheese"
]
for df, factor in zip(cleaned_data, diet_factors):
    df.loc[:, 'Dietary Factor'] = factor  

final_merged_data = pd.concat(cleaned_data, ignore_index=True)
print("Final merged data shape:", final_merged_data.shape)
```
